```python
# Store a person's name in a string variable
name = 'Mike'

# Store a profession in another string variable
profession = "writter"
```

```python
# Store a numeric value (integer) in a variable
books_written = 6
```

```python
# Build a sentence using string concatenation (+)
# Note: books_written is an integer, so we convert it to a string with str()
output = name + " is a " + profession + " and he's written " + str(books_written) + " books."
```

```python
# Build the same sentence using an f-string (cleaner and easier to read)
# f-strings let us place variables directly inside {}
output_2 = f"{name} is a {profession} and he's written {books_written} books."
```

```python
# Print the sentence created with concatenation
print(output)

# Print the sentence created with an f-string
print(output_2)
```
