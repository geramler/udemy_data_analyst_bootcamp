```python
# print() shows text on the screen.
# Strings (text) go inside quotes.
print("Hello World!")
```

```python
# print() can also display numbers directly.
# 32 here is an integer (a whole number).
print(32)
```

```python
# \n creates a new line in the output.
# We convert 32 to text using str(32) so it can be joined with another string.
print("Hello World!\n" + str(32))
```
