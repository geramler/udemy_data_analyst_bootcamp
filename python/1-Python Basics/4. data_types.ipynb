```python
# Integer example: whole numbers without decimals
my_integer = 42

# type() shows the data type of a variable
print(f"My variable is of type: {type(my_integer)}")
```

```python
# Float example: numbers with decimal points
my_float = 3.14

# Confirm the variable's type
print(f"My variable is of type: {type(my_float)}")
```

```python
# String example: text values wrapped in quotes
my_string = 't2.micro'

# Confirm the variable's type
print(f"My variable is of type: {type(my_string)}")
```

```python
# Boolean example: True/False values
is_on = True

# Confirm the variable's type
print(f"My variable is of type: {type(is_on)}")
```
